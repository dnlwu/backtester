```python
import pandas as pd
test_orders = pd.DataFrame(
    [
        [1,'buy','open',pd.to_datetime('2023-04-05'),'IBM'],
        [2,'buy','open',pd.to_datetime('2023-03-08'),'IBM']
    ], 
    index = [1, 2],
    columns=['order_id','order_type','exec_time','exec_date','ticker']
)
test_orders
```

```python
test_orders_mod = test_orders.copy()


test_orders_mod['fwd'] = test_orders.apply(lambda row: get_daily(row['ticker'],start_date=row['exec_date']), axis = 1)
test_orders_mod
```

```python
test_orders.rename(columns={'ticker':'exec_ticker'})
# test_orders.columns.tolist() + ['new']
```

```python
(
    pd.concat(test_orders_mod.set_index(test_orders.columns.tolist()).pop('fwd').to_dict())
    .rename_axis(tuple(test_orders.rename(columns={'ticker':'exec_ticker'}).columns.tolist() + ['new']))
    .reset_index(level=[i for i in range(len(test_orders.columns.tolist()))])
    .reset_index(drop=True)
)
```
